Sebelum menentukan K-Nearest Neighbor kita harus mencari jarak terdekat menggunakan Euclidian Distance:

\begin{align}
d(\mathbf{p},\mathbf{q}) = \sqrt{\sum_{i=1}^n (\mathbf{p}_i - \mathbf{q}_i)^2}
\end{align}

Jadi, kita akan menghitung akar dari penjumlahan (data yang akan diprediksi dikurangi data yang ada (data train) dipangkat 2), label tidak dihitung. 
Kemudian setelah mendapatkan hasilnya kita urutkan dimulai angka terkecil (paling dekat) ke terbesar (paling jauh). Untuk mengetahui hasilnya kita tentukan K (jumlah data yang akan diambil). Misalnya kita pilih K = 5, maka kita ambil 5 data teratas (setelah diurut dari terkecil) kemudian kita hitung jumlah label terbanyak dari 5 data tersebut maka itulah hasilnya.
