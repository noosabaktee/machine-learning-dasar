#Naive Bayes
### Algoritma naive bayes ini menggunakan probabilitas untuk menentukan label sebuah data 

## Rumus:

\begin{align}
P(c \mid X) = \frac{P( x\mid c) \, P(c)}{P(x)}
\end{align}

Untuk lebih mengerti kita langsung praktek saja 
Data diatas memilkki 5 feature (jurusan, gender, school, sks, asisten) dan 1 target (lama studi).

langkah pertama yang barus dilakukan adalah menentukan P(Ci) terlebih dahulu dengan cara menghitung banyak data masing-masing feature dibagi banyak data.
Karena kita memilki 2 feature maka hasilnya menjadi:
\begin{align}
P(LamaStudi = terlambat) = \frac{10}{13} = 0,83
\\[1em]
P(LamaStudi = tepat) = \frac{3}{13} = 0,23
\end{align}

Selanjutnya kita hitung P(X|Ci) dari data diatas dengan cara menghitung jumlah masing-masing data dengan feature tertentu dibagi setiap jumlah feature tersebut(terlambat=13, tepat=3). Contoh:
\begin{align}
P(Jurusan = IPA | LamaStudi = tepat) = \frac{3}{3} = 1
\\
P(Jurusan = IPA | LamaStudi = terlambat) = \frac{4}{10} = 0,4
\\
P(Jurusan = IPS | LamaStudi = tepat) = \frac{0}{3} = 0
\end{align}

keterangan:
* Rumus pertama kita menghitung jumlah jurusan IPA yang memiliki feature = tepat dibagi jumlah feature tepat (3)
* Rumus kedua kita menghitung jumlah jurusan IPA yang memiliki feature = terlambat dibagi jumlah feature terlambat (10)
* Rumus ketiga kita menghitung jumlah jurusan IPS yang memiliki feature = tepat (IPS tidak memiliki data dengan feature = tepat) dibagi jumlah feature terlambat (10)
* dst

Setelah itu kita bisa menghitung probabilitas sebuah data terhadap feature
Misal kita punya data: \\
IPS | WANITA | SURAKARTA | SKS <= 18 | TIDAK | ?

Kita diminta untuk menentukan label yang masih kosong (?).
Maka kita kalikan dengan kategori data uji
\begin{align}
P(X | LamaStudi = tepat * P(LamaStudi = tepat) = (\frac{0}{3} x \frac{1}{3} x \frac{3}{3} x \frac{1}{3} x \frac{1}{3}) x \frac{3}{13} = 0
\\[1em]
P(X | LamaStudi = terlambat * P(LamaStudi = terlambat) = (\frac{5}{10} x \frac{3}{10} x \frac{7}{10} x \frac{3}{10} x \frac{10}{10}) x \frac{10}{13} = 0,024
\end{align}

keterangan:
* Pertama kita menghitung probabilitas untuk feature "tepat" dengan mengalikan seluruh hasil P(X|Ci) data test tepat dengan P(Ci).
Yaitu dengan cara menghitung (Jumlah data IPS yang memiliki feature tepat dibagi jumlah data feature tepat) x (Jumlah data WANITA yang memiliki feature tepat dibagi jumlah data feature tepat) x ... x ... dst kemudian dikali dengan P(Ci) atau jumlah feature tepat dibagi banyaknya data (3/13)
* Kemudian kita menghitung probabilitas untuk feature "terlambat" dengan mengalikan seluruh hasil P(X|Ci) data test terlambat dengan P(Ci).
Yaitu dengan cara menghitung (Jumlah data IPS yang memiliki feature terlambat dibagi jumlah data feature terlambat) x (Jumlah data WANITA yang memiliki feature terlambat dibagi jumlah data feature terlambat) x ... x ... dst kemudian dikali dengan P(Ci) atau jumlah feature terlambat dibagi banyaknya data (10/13)

Langkah terakhir kita lihat hasilnya, karena probabilitas "terlambat" lebih besar daripada "tepat" maka data test tersebut akan memiliki label "terlambat"

## ALERT!
Jika kita lihat, angka nol akan sangat berpengaruh untuk perhitung karena ini menggunakan perkalian. Dari data diatas kita tidak memiliki data IPS dengan feature "tepat" sehingga hasil perhitungan probabilitas menjadi 0, andai saja kita punya 1 data IPS dengan feature "tepat" maka kita tidak akan mendapatkan nilai 0 atau bahkan feature "tepat" bisa lebih tinggi dari feature "terlambat".
Untuk mengatasi masalah tersebut adalah dengan cara "Laplacian Correction".
* pertama kita menambahkan 1 tupel pura-pura ke dalam kelas lamaStudi = tepat. Kenapa? karena lamaStudi = tepat menghasilkan probalitas nol.
* Karena ada 5 feature, maka kita akan menambahkan 5 tupel
* Tupel ini hanya pura-pura, cukup dibayangkan saja. Hanya untuk menghindari probabilitas 0 sehingga tidak berpengaruh terhadap jumlah data asli

\begin{align}
P(X | LamaStudi = tepat * P(LamaStudi = tepat) = (\frac{1}{8} x \frac{2}{8} x \frac{4}{8} x \frac{2}{8} x \frac{2}{8}) x \frac{3}{13} = 0,000225361
\end{align}

Untuk probabilitas angka terakhir (3/13) tidak ditambah dengan 5

Setelah menggunakan Laplacian Correction hasil feature "tetap" masih kecil dari feature "terlambat". 


```python
#Practice
train = ["IPS","WANITA","SURAKARTA","SKS <= 18","TIDAK"]
def NaiveBayes(data):
  target = {}
  n_data = len(df)
  for i in df['lamaStudi'].unique():
    target[i] = len(df.loc[df["lamaStudi"] == i])

  # Mengambil banyaknya data (["IPS","WANITA","SURAKARTA","SKS <= 18","TIDAK"]) berdasarkan target (tepat & terlambat)
  PX = {}
  for k,v in target.items():
    p = []
    for i,j in enumerate(data):
      p.append(len(df.loc[(df[df.columns[i]] == j) & (df["lamaStudi"] == k)]))
    PX[k] = p
    p = []

  result = {}
  for k,v in target.items():
    divide = v #pembagi => banyaknya target(tepat/terlambat)
    # antisipasi nilai nol(0)
    if 0 in PX[k]:
      PX[k] = [x+1 for x in PX[k]]
      divide += len(data)
    r = []
    for i in PX[k]:
      r.append(i/divide)
    # menambahkan banyaknya target (tepat/terlambat) / banyaknya data
    r.append(v/n_data)
    # np.prod digunakan untuk mengalikan semua angka didalam list
    result[k] = round(np.prod(r),8)
    r = []

  print(result)
  # Kesimpulan
  max_value = max(result, key=result.get)
  print(f"Berdasarkan data yang diberikan maka hasilnya adalah {max_value}")
NaiveBayes(train)
```
